Character Text Splitter

```python
with open('../DocumentLoaders/speech.txt') as f:
      text = f.read()
      
print(text)
```

```python
from langchain_text_splitters  import CharacterTextSplitter

textsplitter = CharacterTextSplitter(chunk_size=100, chunk_overlap =90, separator = '\n\n') 
docs = textsplitter.split_text(text)
print(docs)
len(docs)
```
