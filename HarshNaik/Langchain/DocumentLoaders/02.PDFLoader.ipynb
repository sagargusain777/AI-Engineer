PDF Loader In Langchain

```python
from langchain_community.document_loaders import PyPDFLoader


loader = PyPDFLoader("attention.pdf")
document = loader.load()
print(document)
type(document)  #Output : List

final_document = document[0]
print(final_document)
type(final_document) #Ouput : langchain_core.documents.base.Document
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("attention.pdf")
documents = loader.load()
print(documents)
type(documents)   #Output : List
final_docs = documents[0]
print(final_docs)
type(final_docs) #Output : langchain_core.documents.base.Document
docsdata = final_docs.metadata
print(docsdata)
```
