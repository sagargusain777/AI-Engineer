What is parse_only?
üîπ Problem it solves

Web pages are huge:

headers

footers

navigation bars

ads

scripts

styles

Most of the time, you only want specific parts of the page.

parse_only tells BeautifulSoup:

‚ùù Parse ONLY these parts of the HTML ‚Äî ignore everything else ‚ùû

This makes:

parsing faster

extracted text cleaner

embeddings better

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(web_paths=["https://langchain.com",],
bs_kwargs =dict(
    parse_only = bs4.SoupStrainer(
        class_=("main_wrapper","section-logos")
    )
))
documents = loader.load()
print(documents)
```

Why LangChain uses dictionary so much

LangChain passes options, not fixed values.

Think:

‚ÄúI want to tell the tool what this value means‚Äù

Example:

bs_kwargs = {
    "parse_only": something
}


This clearly tells LangChain:

key = "parse_only" ‚Üí what setting

value = something ‚Üí how to apply it

our LangChain code rewritten VERY simply
Your original code
bs_kwargs = dict(
    parse_only = bs4.SoupStrainer(
        class_=("main_wrapper", "section-logos")
    )
)

Same thing, but easier to read
bs_kwargs = {
    "parse_only": bs4.SoupStrainer(
        class_=("main_wrapper", "section-logos")
    )
}


üí° dict(...) and {} are the same thing

What LangChain does with this dictionary

Inside WebBaseLoader, something like this happens:

BeautifulSoup(html, "html.parser", **bs_kwargs)


Your dictionary becomes:

BeautifulSoup(
    html,
    "html.parser",
    parse_only=SoupStrainer(...)
)


So:

dictionary ‚Üí unpacked ‚Üí keyword arguments
